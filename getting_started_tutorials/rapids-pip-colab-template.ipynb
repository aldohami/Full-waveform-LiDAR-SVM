```
import os
import re
import numpy as np
import pandas as pd
from plyfile import PlyData
import json

# ----- Helper function to remove scale suffix from field names -----
def remove_scale_suffix(name):
    # Remove underscore and any parenthesized number at the end, e.g., _(...).
    return re.sub(r'_\(.*\)$', '', name)

# ----- Load the class dictionary (labels mapping) -----
# Update this path if your JSON file is in a different location
class_dict_path = "/content/class_dict.json"
with open(class_dict_path, "r") as f:
    class_dict = json.load(f)

# ----- Functions to load and merge a single scan -----
def load_ply_as_df(ply_path, fields_to_keep=None, rename_scale=False):
    """
    Load a PLY file and return its vertex data as a pandas DataFrame.
    If rename_scale is True, remove any scale suffix from the field names.
    If fields_to_keep is provided, only those columns (after renaming) are kept.
    """
    plydata = PlyData.read(ply_path)
    vertex_data = plydata['vertex'].data
    df = pd.DataFrame(vertex_data)
    if rename_scale:
        df.columns = [remove_scale_suffix(c) for c in df.columns]
    if fields_to_keep is not None:
        available = [f for f in fields_to_keep if f in df.columns]
        df = df[available]
    return df

def load_fwf_npy(npy_path):
    """
    Load a NumPy file containing FWF features.
    Assumes the first row is headers (feature names) and the rest are numeric values.
    Also removes any scale suffix from the headers.
    """
    arr = np.load(npy_path, allow_pickle=True)
    if isinstance(arr[0][0], str):
        headers = np.array([remove_scale_suffix(h) for h in arr[0]])
        data = arr[1:]
    else:
        num_features = arr.shape[1]
        headers = np.array([f"fwf_feature_{i}" for i in range(num_features)])
        data = arr
    return headers, data

def merge_scan(ply_path, npy_path, ply_fields_to_keep=None, rename_scale=False):
    """
    Merges the geometric data from a PLY file with FWF features from a NumPy file.
    Returns a DataFrame containing only the specified PLY columns and all FWF features.
    """
    df_ply = load_ply_as_df(ply_path, fields_to_keep=ply_fields_to_keep, rename_scale=rename_scale)
    fwf_headers, fwf_data = load_fwf_npy(npy_path)

    # Check that the number of points match
    if df_ply.shape[0] != fwf_data.shape[0]:
        raise ValueError(f"Row mismatch: PLY has {df_ply.shape[0]} rows but FWF file has {fwf_data.shape[0]} rows.")

    df_fwf = pd.DataFrame(fwf_data, columns=fwf_headers)
    # Merge side-by-side (resetting indices for proper alignment)
    df_merged = pd.concat([df_ply.reset_index(drop=True), df_fwf.reset_index(drop=True)], axis=1)
    return df_merged

def load_dataset(dataset_folder, ply_fields_to_keep=None, rename_scale=False):
    """
    Expects the folder structure:
        dataset_folder/
            Geometric/  --> contains PLY files
            FWF/        --> contains corresponding FWF .npy files
    For each scan, it merges the PLY and FWF files (matched by scan ID) and
    concatenates them into one big DataFrame.
    """
    geo_folder = os.path.join(dataset_folder, "Geometric")
    fwf_folder = os.path.join(dataset_folder, "FWF")

    merged_scans = []
    for file in os.listdir(geo_folder):
        if file.endswith(".ply"):
            # Example PLY filename: "2024-04-05_FW_Westbahnhof_04_pointcloud - Cloud.ply"
            base = os.path.splitext(file)[0]
            # Extract the scan ID (e.g., "2024-04-05_FW_Westbahnhof_04")
            scan_id = base.split("_pointcloud")[0]
            # Construct the corresponding FWF filename
            npy_file = f"{scan_id}_waveform_processed.npy"
            ply_path = os.path.join(geo_folder, file)
            npy_path = os.path.join(fwf_folder, npy_file)
            if os.path.exists(npy_path):
                try:
                    df_scan = merge_scan(ply_path, npy_path, ply_fields_to_keep=ply_fields_to_keep, rename_scale=rename_scale)
                    merged_scans.append(df_scan)
                    print(f"Loaded and merged scan: {scan_id}")
                except Exception as e:
                    print(f"Error merging scan {scan_id}: {e}")
            else:
                print(f"Matching FWF file not found for scan {scan_id}")

    if merged_scans:
        df_dataset = pd.concat(merged_scans, ignore_index=True)
        return df_dataset
    else:
        return pd.DataFrame()

# ----- Function to fill NaN values with 0 -----
def fill_nan_with_zero(df):
    return df.fillna(0)

# ----- Main execution: Load Train and Test datasets -----
# In your screenshot, you appear to have "train_data" and "test_data" folders.
# If your folders are named "train_data/Geometric" + "train_data/FWF" and so on, do this:
train_folder = "/content/Train"
test_folder  = "/content/Test"

# Define common PLY fields to keep (without scale suffixes).
# Update as needed for your actual columns.
ply_fields_to_keep = [
    'scalar_riegl_reflectance', 'scalar_riegl_amplitude', 'scalar_riegl_deviation',
    'red', 'green', 'blue',
    'scalar_Roughness', 'scalar_1st_order_moment', 'scalar_Mean_curvature', 'scalar_Gaussian_curvature',
    'scalar_Normal_change_rate', 'scalar_Number_of_neighbors', 'scalar_Surface_density', 'scalar_Volume_density',
    'scalar_Eigenvalues_sum', 'scalar_Omnivariance', 'scalar_Eigenentropy', 'scalar_Anisotropy', 'scalar_Planarity',
    'scalar_Linearity', 'scalar_PCA1', 'scalar_PCA2', 'scalar_Surface_variation', 'scalar_Sphericity',
    'scalar_Verticality', 'scalar_1st_eigenvalue', 'scalar_2nd_eigenvalue', 'scalar_3rd_eigenvalue',
    'scalar_labels_0', 'scalar_labels_1', 'scalar_labels_3'
]

df_train = load_dataset(train_folder, ply_fields_to_keep=ply_fields_to_keep, rename_scale=True)
df_test  = load_dataset(test_folder, ply_fields_to_keep=ply_fields_to_keep, rename_scale=True)

# Fill NaN values with 0
df_train = fill_nan_with_zero(df_train)
df_test  = fill_nan_with_zero(df_test)

print("Train dataset shape:", df_train.shape)
print("Test dataset shape:", df_test.shape)

print("\nColumns in Train dataset:")
print(df_train.columns.tolist())
print("\nFirst 10 rows of Train dataset:")
print(df_train.head(10))
```

```
# =======================
# Colab Example with cuML
# =======================

import numpy as np
import pandas as pd
from cuml.svm import SVC  # RAPIDS cuML SVM (GPU-accelerated)
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
import json

# ----- Load the class dictionary (labels mapping) -----
# (Update this path or remove if you don't need the JSON dictionary)
class_dict_path = "/content/class_dict.json"
with open(class_dict_path, "r") as f:
    class_dict = json.load(f)

# ----- Define feature sets with scale-stripped names -----
FEATURE_SETS = {
    "geom_only": [
        'scalar_Roughness',
        'scalar_1st_order_moment',
        'scalar_Mean_curvature',
        'scalar_Gaussian_curvature',
        'scalar_Normal_change_rate',
        'scalar_Number_of_neighbors',
        'scalar_Surface_density',
        'scalar_Volume_density',
        'scalar_Eigenvalues_sum',
        'scalar_Omnivariance',
        'scalar_Eigenentropy',
        'scalar_Anisotropy',
        'scalar_Planarity',
        'scalar_Linearity',
        'scalar_PCA1',
        'scalar_PCA2',
        'scalar_Surface_variation',
        'scalar_Sphericity',
        'scalar_Verticality',
        'scalar_1st_eigenvalue',
        'scalar_2nd_eigenvalue',
        'scalar_3rd_eigenvalue'
    ],
    "fwf_only":  [
        "total_amplitude", "weighted_mu", "mean_sigma", "total_area",
        "num_gaussians", "r_squared", "average_inter_peak_distance",
        "normalized_prominence", "echo_asymmetry", "rise_time", "decay_time"
    ],
    "geom_plus_fwf": [
        'scalar_Roughness',
        'scalar_1st_order_moment',
        'scalar_Mean_curvature',
        'scalar_Gaussian_curvature',
        'scalar_Normal_change_rate',
        'scalar_Number_of_neighbors',
        'scalar_Surface_density',
        'scalar_Volume_density',
        'scalar_Eigenvalues_sum',
        'scalar_Omnivariance',
        'scalar_Eigenentropy',
        'scalar_Anisotropy',
        'scalar_Planarity',
        'scalar_Linearity',
        'scalar_PCA1',
        'scalar_PCA2',
        'scalar_Surface_variation',
        'scalar_Sphericity',
        'scalar_Verticality',
        'scalar_1st_eigenvalue',
        'scalar_2nd_eigenvalue',
        'scalar_3rd_eigenvalue',
        "total_amplitude", "weighted_mu", "mean_sigma", "total_area",
        "num_gaussians", "r_squared", "average_inter_peak_distance",
        "normalized_prominence", "echo_asymmetry", "rise_time", "decay_time"
    ],
    "geom_plus_fwf_plus_color": [
        'scalar_Roughness',
        'scalar_1st_order_moment',
        'scalar_Mean_curvature',
        'scalar_Gaussian_curvature',
        'scalar_Normal_change_rate',
        'scalar_Number_of_neighbors',
        'scalar_Surface_density',
        'scalar_Volume_density',
        'scalar_Eigenvalues_sum',
        'scalar_Omnivariance',
        'scalar_Eigenentropy',
        'scalar_Anisotropy',
        'scalar_Planarity',
        'scalar_Linearity',
        'scalar_PCA1',
        'scalar_PCA2',
        'scalar_Surface_variation',
        'scalar_Sphericity',
        'scalar_Verticality',
        'scalar_1st_eigenvalue',
        'scalar_2nd_eigenvalue',
        'scalar_3rd_eigenvalue',
        "red", "green", "blue",
        "total_amplitude", "weighted_mu", "mean_sigma", "total_area",
        "num_gaussians", "r_squared", "average_inter_peak_distance",
        "normalized_prominence", "echo_asymmetry", "rise_time", "decay_time"
    ]
}

# ----- Define label columns -----
LABELS = ["scalar_labels_0", "scalar_labels_1", "scalar_labels_3"]

# ----- df_train and df_test should already exist in your Colab environment -----
# For example, from your merging code. Let's assume you have something like:
# df_train = ...
# df_test  = ...

print("Train dataset shape:", df_train.shape)
print("Test dataset shape:", df_test.shape)
print("Columns in Train dataset:")
print(df_train.columns.tolist())

results_summary = []

# Loop over each label and each feature set with progress bars
for label in tqdm(LABELS, desc="Labels"):
    mapping_key = label.replace("scalar_", "")
    if mapping_key not in class_dict:
        print(f"Mapping for {label} not found in JSON; skipping this label.")
        continue
    label_mapping = class_dict[mapping_key]
    print(f"\nUsing label column: {label}")
    print("Class mapping:", label_mapping)

    if label not in df_train.columns or label not in df_test.columns:
        print(f"Label column {label} not found in Train or Test data; skipping.")
        continue

    for feat_set_name, feat_list in tqdm(FEATURE_SETS.items(), desc="Feature Sets", leave=False):
        missing_features = [f for f in feat_list if f not in df_train.columns]
        if missing_features:
            print(f"For feature set '{feat_set_name}', missing features: {missing_features}. Skipping this set.")
            continue

        # Extract features and labels from Train and Test
        X_train = df_train[feat_list].values.astype(np.float32)
        y_train = df_train[label].values.astype(np.int32)
        X_test = df_test[feat_list].values.astype(np.float32)
        y_test = df_test[label].values.astype(np.int32)

        # Downsample for speed if needed
        downsample_frac = 0.1
        train_idx = np.random.choice(len(X_train), int(len(X_train) * downsample_frac), replace=False)
        test_idx = np.random.choice(len(X_test), int(len(X_test) * downsample_frac), replace=False)
        X_train, y_train = X_train[train_idx], y_train[train_idx]
        X_test, y_test = X_test[test_idx], y_test[test_idx]
        print(f"Downsampled shapes - X_train: {X_train.shape}, X_test: {X_test.shape}")

        # Standardize features (fit on train, transform both)
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        # Train cuML SVC (GPU-accelerated if you have RAPIDS installed)
        clf = SVC(kernel='rbf', C=1.0, random_state=42)
        clf.fit(X_train_scaled, y_train)
        y_pred = clf.predict(X_test_scaled)

        acc = accuracy_score(y_test, y_pred)
        report = classification_report(y_test, y_pred, output_dict=True)
        print(f"Label: {label}, Feature set: {feat_set_name}, Accuracy: {acc:.4f}")

        results_summary.append({
            "label": label,
            "label_mapping": label_mapping,
            "feature_set": feat_set_name,
            "accuracy": acc,
            "classification_report": report
        })

# Convert summary to a DataFrame for further analysis
results_df = pd.DataFrame(results_summary)
print("\nSummary of SVM classification experiments:")
print(results_df)
```
